```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Life_Expectancy_00_15.csv', sep = ';')
br = df.loc[df['Country']=='Brazil']
ar = df.loc[df['Country']== 'Argentina']
bo = df.loc[df['Country']== 'Bolivia']
ch = df.loc[df['Country']== 'Chile']
df_2 = pd.concat([br, ar, bo, ch])
df_2.head()
```

```python
#Boxplot Expectativa de vida 
df_2.boxplot(column=['Life Expectancy'])
plt.title('Media de Expectativa de vida')
plt.figure(figsize = (10,8))
plt.show()

br.boxplot(column=['Life Expectancy'])
plt.title('Expectativa de Vida Brasileira')
plt.figure(figsize = (10,8))
plt.show()

ar.boxplot(column=['Life Expectancy'])
plt.title('Expectativa de Vida Argentina')
plt.figure(figsize = (10,8))
plt.show()

bo.boxplot(column=['Life Expectancy'])
plt.title('Expectativa de vida Boliviana')
plt.figure(figsize = (10,8))
plt.show()

ch.boxplot(column=['Life Expectancy'])
plt.title('Expectativa de Vida do Chile')
plt.figure(figsize = (10,8))
plt.show()
```

```python
#Consumo de bebidas Anual
br.boxplot(column=['Beer consumption per capita'])
plt.title('Consumo de cerveja')
```

```python
#Obesidade entre adultos
df_2.boxplot(column=['Obesity among adults'])
plt.title('Obesidade entre Adultos')
```

```python
plt.figure(figsize = (10,8))
grafico = plt.scatter(
    br['Life Expectancy'], 
    br['Year'], 
    c='red')
plt.title('Espectativa de Vida Brasileira')
plt.xlabel("Espectativa de Vida")
plt.ylabel("Ano")
plt.plot()

plt.figure(figsize = (10,8))
grafico = plt.scatter(
    ar['Life Expectancy'], 
    ar['Year'], 
    c='red')
plt.title('Espectativa de Vida Argentina')
plt.xlabel("Espectativa de Vida")
plt.ylabel("Ano")
plt.plot()

plt.figure(figsize = (10,8))
grafico = plt.scatter(
    bo['Life Expectancy'], 
    bo['Year'], 
    c='red')
plt.title('Espectativa de Vida Boliviana')
plt.xlabel("Espectativa de Vida")
plt.ylabel("Ano")
plt.plot()

plt.figure(figsize = (10,8))
grafico = plt.scatter(
    ch['Life Expectancy'], 
    ch['Year'], 
    c='red')
plt.title('Espectativa de Vida Chilena')
plt.xlabel("Espectativa de Vida")
plt.ylabel("Ano")
plt.plot()
```

```python
espec = df_2['Life Expectancy']
basico = df_2['Year']
widht_n = 0.8
plt.figure(figsize=(10, 8))
plt.axis(ymax=90)
plt.bar(basico, espec, width=widht_n)
plt.title('Espectativa de vida dos anos de 2000 à 2015')
plt.xlabel('Ano')
plt.ylabel('Espectativa de vida')
plt.show()
```
